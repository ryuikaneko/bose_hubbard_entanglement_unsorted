```python
import numpy as np
import matplotlib.pyplot as plt
from numba import jit
import time
```

```python
# https://github.com/XanaduAI/thewalrus/blob/master/thewalrus/_permanent.py

@jit(nopython=True)
def perm_bbfg(M):  # pragma: no cover
    """
    Returns the permanent of a matrix using the bbfg formula in Gray ordering
    The code is a re-implementation from a Python 2 code found in
    `Permanent code golf
    <https://codegolf.stackexchange.com/questions/97060/calculate-the-permanent-as-quickly-as-possible>`_
    using Numba.
    Args:
        M (array) : a square array.
    Returns:
        float or complex: the permanent of a matrix ``M``
    """

    n = len(M)
    if n == 0:
        return M.dtype.type(1.0)
    row_comb = np.sum(M, 0)
    total = 0
    old_gray = 0
    sign = +1
    binary_power_dict = np.array([2**i for i in range(n)])
    num_loops = 2 ** (n - 1)
    for bin_index in range(1, num_loops + 1):
        reduced = np.prod(row_comb)
        total += sign * reduced
        new_gray = bin_index ^ (bin_index // 2)
        gray_diff = old_gray ^ new_gray
        gray_diff_index = np.searchsorted(binary_power_dict,gray_diff)
        new_vector = M[gray_diff_index]
        direction = 2 * ((old_gray > new_gray) - (old_gray < new_gray))
        for i in range(n):
            row_comb[i] += new_vector[i] * direction
        sign = -sign
        old_gray = new_gray
    return total / num_loops
```

```python
@jit(nopython=True)
def calc_rsks(L):
    rs = np.arange(1,L+1) ## from 1 to L
    ks = np.linspace(np.pi/(L+1.0),L*np.pi/(L+1.0),L)
    return rs,ks

@jit(nopython=True)
def calc_eks(J,ks):
    return -2.0*J*np.cos(ks)

@jit(nopython=True)
def calc_xkrs(L,ks,rs):
    return np.sqrt(2.0/(L+1.0))*np.sin(np.outer(ks,rs))

@jit(nopython=True)
def calc_ys(t,eks,xkrs):
    xc = (1.0+0.0j)*xkrs # for numba
    expeks = np.diag(np.exp(1j*t*eks))
    return xc @ expeks @ xc

@jit(nopython=True)
def calc_fs(ys,state):
    diag = (1.0+0.0j)*np.diag(state)
    return ys @ diag @ ys.T.conjugate()

@jit(nopython=True)
def calc_z(L,state,t,eks,xkrs):
    ys = calc_ys(t,eks,xkrs)
    fs = calc_fs(ys,state)
    dlt = np.eye(L,dtype=np.complex128)
    return np.vstack( (np.hstack((fs,dlt-fs)), np.hstack((dlt-fs,fs))) )

@jit(nopython=True)
def get_dat(L,state,J,ts):
    rs, ks = calc_rsks(L)
    eks = calc_eks(J,ks)
    xkrs = calc_xkrs(L,ks,rs)
    data = np.zeros((len(ts),2),dtype=np.float64)
    for i,t in enumerate(ts):
        matz = calc_z(L,state,t,eks,xkrs)
        data[i] = np.array([t,-np.log(perm_bbfg(matz).real)])
    return data
```

```python
J = 1.0
ts = np.linspace(0,20,201)
L = 6
L_As = np.arange(0,L//2+1)

datas = np.zeros((len(L_As),len(ts),2),dtype=np.float64)
for i,L_A in enumerate(L_As):
    start = time.time()
    state = np.array([1.0 if i<L_A else 0.0 for i in range(L)])
    datas[i] = get_dat(L,state,J,ts)
    end = time.time()
    print("L,L_A,calctime:",L,L_A,end-start)
```

```python
for i,L_A in enumerate(L_As):
    plt.plot(datas[i,:,0],datas[i,:,1])
plt.show()
```

```python
# https://doi.org/10.1126/science.aaf6725

# when #A=1 and #B=1 --> combined_volume = 1+1 = 2

J = 1.0
ts = np.linspace(0,20,201)
L = 6
rs = np.arange(1,L+1) ## from 1 to L

state_A = np.array([1.0 if r==1 else 0.0 for r in rs])
data_A = get_dat(L,state_A,J,ts)
print(state_A)
state_B = np.array([1.0 if r==6 else 0.0 for r in rs])
print(state_B)
data_B = get_dat(L,state_B,J,ts)
#state_AB = np.array([1.0 if 1<=r<=3 else 0.0 for r in rs])
state_AB = np.array([1.0 if r==1 or r==6 else 0.0 for r in rs])
print(state_AB)
data_AB = get_dat(L,state_AB,J,ts)

mi = np.zeros((len(ts),2),dtype=np.float64)
mi[:,0] = ts[:]
mi[:,1] = data_A[:,1] + data_B[:,1] - data_AB[:,1]

plt.plot(data_A[:,0],data_A[:,1])
plt.plot(data_B[:,0],data_B[:,1])
plt.plot(data_AB[:,0],data_AB[:,1])
plt.show()

plt.plot(mi[:,0],mi[:,1])
plt.show()
```
